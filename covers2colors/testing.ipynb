```python
from covers2colors import CoverColors
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
covercolors = CoverColors('closure in moscow', "pink lemonade")
cmap = covercolors.generate_cmap(n_colors=3, random_state=42)

colors = cmap.colors
with plt.style.context("dark_background"):
    for i, color in enumerate(colors):
        plt.plot(range(10), [_+i+1 for _ in range(10)], color=color, linewidth=5)
plt.show()

for i, color in enumerate(colors):
    plt.plot(range(10), [_+i+1 for _ in range(10)], color=color, linewidth=5)
plt.show()
```

```python
covercolors.display_with_colorbar(cmap)
```

```python
cmaps, best_n_colors, ssd = covercolors.generate_optimal_cmap(max_colors=10, random_state=42)
best_cmap = cmaps[best_n_colors]
colors = best_cmap.colors

with plt.style.context("dark_background"):
    for i, color in enumerate(colors):
        plt.plot(range(10), [_+i+1 for _ in range(10)], color=color, linewidth=5)
plt.show()

for i, color in enumerate(colors):
    plt.plot(range(10), [_+i+1 for _ in range(10)], color=color, linewidth=5)
plt.show()

covercolors.display_with_colorbar(best_cmap)
```
